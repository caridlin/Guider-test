```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.pyplot as plt
import numpy as np

Array = image_data
x = int(0.5 * image_data.shape[1])
y = int(0.5 * image_data.shape[0])

fig_mpl, ax = plt.subplots(facecolor = 'white')

line = ax.imshow(Array, aspect='equal')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size = "5%", pad = 0.05)
caxb = divider.append_axes("bottom", size = "30%", pad = 0.05)
caxl = divider.append_axes("left", size = "20%", pad = 0.05)

bar = fig_mpl.colorbar(line, cax = cax, orientation = 'vertical') 
ax.axhline(y, color = 'red')
ax.axvline(x, color = 'red')      

X = np.linspace(0, 2 * x - 1 , 2 * x)
Y = np.linspace(0, 2 * y - 1 , 2 * y)
projb, = caxb.plot(X, Array[int(x)], color = 'red')
projl, = caxl.plot(Array[:, int(y)], Y, color = 'red')

caxb.set_ylim(-0.1*np.max(Array), 1.1*np.max(Array))
caxb.set_xlim(0, x * 2  - 1)
caxl.set_xlim(-0.1*np.max(Array), 1.1*np.max(Array))
caxl.set_ylim(0, y * 2 - 1)

ax.set_xticks([])
ax.set_yticks([])
caxb.set_xticks([])
caxl.set_yticks([])
caxb.set_yticks([np.min(Array), np.max(Array)])
caxl.set_xticks([np.min(Array), np.max(Array)])
caxb.yaxis.tick_right()

for tick in caxl.get_xticklabels():
    tick.set_rotation(-90)

caxb.grid(color = 'black', marker = 8)
caxl.grid(color = 'black', marker = 8)

fig_mpl.subplots_adjust(wspace = 0)
fig_mpl.tight_layout()
```
